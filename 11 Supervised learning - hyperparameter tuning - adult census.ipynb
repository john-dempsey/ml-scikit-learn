```python
import pandas as pd

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_validate
```

```python
adult_census = pd.read_csv("../datasets/adult-census.csv")
adult_census.head()
```

```python
numerical_columns = ["age", "capital-gain", "capital-loss", "hours-per-week"]
target_column = "class"

X = adult_census[numerical_columns]
y = adult_census[target_column]
```

```python
pipeline = Pipeline(
    steps=[
        ("preprocessor", StandardScaler()),
        ("classifier", LogisticRegression()),
    ]
)
```

```python
for parameter in pipeline.get_params():
    print(parameter)
```

```python
for C in [0.001, 0.01, 0.1, 1, 10, 100]:
    pipeline.set_params(classifier__C=C)
    cv_results = cross_validate(pipeline, X, y)
    scores = cv_results["test_score"]
    print(
        f"Accuracy score via cross-validation with C={C}:\n"
        f"{scores.mean():.3f} ± {scores.std():.3f}"
    )
```
