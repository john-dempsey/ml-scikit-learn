```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
penguins = pd.read_csv("datasets/penguins_regression.csv")
feature_column = "Flipper Length (mm)"
target_column = "Body Mass (g)"
X, y = penguins[[feature_column]], penguins[target_column]
```

```python
linear_regression = LinearRegression()
linear_regression.fit(X, y)
```

```python
weight_flipper_length = linear_regression.coef_[0]
weight_flipper_length
```

```python
intercept_body_mass = linear_regression.intercept_
intercept_body_mass
```

```python
flipper_length_range = np.linspace(X.min(), X.max(), num=300)
predicted_body_mass = (
    weight_flipper_length * flipper_length_range + intercept_body_mass
)
```

```python
sns.scatterplot(x=X[feature_column], y=y, color="black", alpha=0.5)
plt.plot(flipper_length_range, predicted_body_mass)
_ = plt.title("Model using LinearRegression from scikit-learn")
```

```python
inferred_body_mass = linear_regression.predict(X)
model_error = mean_squared_error(y, inferred_body_mass)
print(f"The mean squared error of the optimal model is {model_error:.2f}")
```

```python
model_error = mean_absolute_error(y, inferred_body_mass)
print(f"The mean absolute error of the optimal model is {model_error:.2f} g")
```
